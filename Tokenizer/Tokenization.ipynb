Token(토큰) / Tokenization(토큰화) / Tokenizer (토큰하는 방식)

Token(토큰) 

'나는 책상 위에 사과를 먹었다' -> '나는', '책상', '위에', '사과를', '먹었다'

Tokenization(토큰화) 

'나는 책상 위에 사과를 먹었다' ->  {'나는': 0, '책상': 1, '위에': 2, '사과를': 3, '먹었다': 4, '알고': 5, '보니': 6, '그': 7, '사과는': 8, 'Jason': 9, '것이었다': 10, '그래서': 11, 'Jason에게': 12, '했다': 13}

Tokenizer (토큰화하는 방식)

- Basic - 띄어쓰기 
        - 새로운 단어 <Unk>

- BPE - 1.Character Based Tokenization
        2.N-gram
        3.BPE(Byte Pair Encoding)

Basic

```python
S1 = '나는 책상 위에 사과를 먹었다'
S2 = '알고 보니 그 사과는 Jason 것이었다'
S3 = '그래서 Jason에게 사과를 했다'
```

```python
print(S1.split())
print(S2.split())
print(S3.split())
```

```python
print(list(S1))
```

```python
token2idx={}
index=0

for sentence in [S1,S2,S3]:
    tokens =sentence.split()
    for token in tokens:
        if token2idx.get(token) == None:
            token2idx[token]=index
            index+=1
            
print(token2idx)
```

```python
def indexed_sentence(sentence):
    return [token2idx[token] for token in sentence]

S1_i = indexed_sentence(S1.split())
print(S1_i)

S2_i = indexed_sentence(S2.split())
print(S2_i)

S3_i = indexed_sentence(S3.split())
print(S3_i)
```

OOV 처리를 위한 Unknown 토큰 

```python
S4='나는 책상 위에 배를 먹었다'
indexed_sentence(S4.split())
```

```python
# 기존 단어 사전에 저장된 단어들의 인덱스를 한 칸 뒤로 보내기
token2idx = {t: i+1 for t,i in token2idx.items()}
# <unk>을 0으로 지정 
token2idx['<unk>'] = 0 

#  token이 없을 경우 <unk> token의 0으로 치환
def indexed_sentence_unk(sentence):
    return [token2idx.get(token,token2idx['<unk>']) for token in sentence]
    # return [token2idx.get(token,0]) for token in sentence]  # <unk>을 0으로 지정해놓았기 때문에 같음 
    
indexed_sentence_unk(S4.split())
```
